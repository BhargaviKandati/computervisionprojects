```python
import cv2 as cv
from matplotlib import pyplot as pt
```

```python
image = cv.imread('./test.jpg')
pt.imshow(image)
```

```python
# computer will always read image as BGR 
def showImage(image=None,title=""):
    image = cv.cvtColor(image,cv.COLOR_BGR2RGB)
    pt.title(title)
    pt.imshow(image)

image = cv.imread('./test.jpg')
showImage(image,"Elephant")
```

```python
image = cv.imread('test.jpg')
new_image = cv.cvtColor(image,cv.COLOR_BGR2GRAY)
showImage(new_image,"Elephant")
```

```python
image = cv.imread('test.jpg')
new_image = cv.flip(image,1)
showImage(new_image,"ELEPHANT")
```
